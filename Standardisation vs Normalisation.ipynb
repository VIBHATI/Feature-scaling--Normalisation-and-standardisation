```python
import numpy as np # linear algebra
import pandas as pd # data processing
import matplotlib.pyplot as plt # data visualization
```

Generating data sampled from normal, poisson and uniform distribution and visualizing the effect of applying MinMaxScaler and StandardScaler on data distribution. Normal or gaussian distributions are observed in the real world in several scneraios including height of the population, Income distribution in an economy. Poisson distribution is a skewed distribution and is typically observed in retail sales data. When we roll a die or toss an unbiased coin the observations follow a uniform distribution. 

```python

lambd=1
size=100

mean=2
sd=.75
# poisson distribution
p=np.random.poisson(lambd,size)
# normal distribution
n = np.random.normal(mean, sd, size)
# uniform distribution
u=np.random.uniform(0,1,3000)
```

```python
# Feature scaling
from sklearn.preprocessing import MinMaxScaler,normalize,StandardScaler,minmax_scale

min_max=MinMaxScaler()
n_mm=min_max.fit_transform(n.reshape(-1,1))

standr=StandardScaler()
n_sc=standr.fit_transform(n.reshape(-1,1))

min_max=MinMaxScaler()
p_mm=min_max.fit_transform(p.reshape(-1,1))

standr=StandardScaler()
p_sc=standr.fit_transform(p.reshape(-1,1))

min_max=MinMaxScaler()
u_mm=min_max.fit_transform(u.reshape(-1,1))

standr=StandardScaler()
u_sc=standr.fit_transform(u.reshape(-1,1))

```

```python
import seaborn as sns
sns.set(color_codes=True)

sns.distplot(p)
sns.distplot(n)
sns.distplot(u)
#sns.distplot(p_sc)
```

```python
sns.kdeplot(n, shade=True,label='Normal')
sns.kdeplot(p, shade=True,label='Poisson')
sns.kdeplot(u, shade=True,label='Uniform')
plt.legend()
```

```python
sns.kdeplot(n, shade=True,label='Normal data')
sns.kdeplot(n_sc.reshape(100,), shade=True,label='StandardScaled')
sns.kdeplot(n_mm.reshape(100,), shade=True,label='MinMaxScaler')
```

```python
sns.kdeplot(p, shade=True,label='Poisson data')
sns.kdeplot(p_sc.reshape(100,), shade=True,label='StandardScaled')
sns.kdeplot(p_mm.reshape(100,), shade=True,label='MinMaxScaler')
```

```python
sns.kdeplot(u, shade=True,label='Uniform data',color='black')
sns.kdeplot(u_sc.reshape(3000,), shade=True,label='StandardScaled',color='green')
sns.kdeplot(u_mm.reshape(3000,), shade=True,label='MinMaxScaler',color='yellow')
```

Here we observe that there is no effect of min-max scaler on the uniform distribution ranging form 0 to 1
